```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page=1
every_quote=[]
while(True):
  url=f"https://quotes.toscrape.com/tag/life/page/{page}/"
  response= requests.get(url)
  if response.status_code != 200:
     break 
  soup=BeautifulSoup(response.text,"lxml")
  quotesdiv=soup.find_all("div", class_="quote")
  if len(quotesdiv) == 0:
     break
  for quotediv in quotesdiv:
     quote=quotediv.find("span",class_="text").get_text()
     author=quotediv.find("small").get_text()
     every_quote.append([author,quote])
  page+=1
```

```python
df=pd.DataFrame(every_quote,columns=["Author","Quotes"])
```

```python
df.to_csv("data.csv",index=False)
```
